```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv('facct2026_annotation1.csv')
data2 = pd.read_csv('facct2026_annotation2.csv')

data1 = data1.rename(columns={"Ad type": "Ad type - Model", "Ad topic": "Ad topic - Model"})
data1 = data1[["Ad type - Model", "Ad topic - Model", "Ad type - Annotation", "Ad topic - Annotation"]]
data2 = data2.rename(columns={"Ad type": "Ad type - Model", "Ad topic": "Ad topic - Model"})
data2 = data2[["Ad type - Model", "Ad topic - Model", "Ad type - Annotation", "Ad topic - Annotation"]]
```

```python
ct = pd.crosstab(data1['Ad type - Model'], data1['Ad type - Annotation'], dropna=False)
ct = ct.reindex(index=['formal', 'influencer', 'non-ad', 'other'], columns=['formal', 'influencer', 'non-ad', 'other'], fill_value=0)
label_map = {'influencer': 'disclosed', 'non-ad': 'undisclosed'}
ct.index = [label_map.get(label, label) for label in ct.index]
ct.columns = [label_map.get(label, label) for label in ct.columns]
print('Annotator1: Ad type')
display(ct)
plt.figure(figsize=(5, 4))
plt.title('Annotator1: Ad type')
plt.imshow(ct.values, cmap='Blues')
plt.xticks(range(len(ct.columns)), ct.columns, rotation=45, ha='right')
plt.yticks(range(len(ct.index)), ct.index)
plt.xlabel('Annotator')
plt.ylabel('Model')
for i in range(ct.shape[0]):
    for j in range(ct.shape[1]):
        plt.text(j, i, str(ct.iat[i, j]), ha='center', va='center', color='black')
plt.colorbar()
plt.tight_layout()
plt.show()
```
